```python
import matplotlib.pyplot as plt
import numpy as np

def budget_constraint(p1, p2, budget):
    # Generate range of quantities for good 1
    q1 = np.linspace(0, budget/p1, 100)

    # Calculate the corresponding quantity for good 2
    q2 = (budget - p1 * q1) / p2

    return q1, q2

# Define the prices and budget
price1 = 1.98
price2 = 1
budget = 2000

# Calculate the quantities for the budget constraint
q1, q2 = budget_constraint(price1, price2, budget)

# Plot the budget constraint
plt.figure(figsize=(6, 6))
plt.plot(q1, q2)
plt.xlabel('Quantity of Good 1')
plt.ylabel('Quantity of Good 2')
plt.title('Budget Constraint')
plt.grid(True)
plt.xlim(0, budget/price1)
plt.ylim(0, budget/price2)
plt.show()
```

```python
def budsjett(M, p1, p2, x1):
    return M/p2 - p1*x1/p2

x=np.linspace(0.001,100, 100)

fig, ax = plt.subplots()
ax.set_ylabel('$x_2$', loc='top', rotation=0)
ax.set_xlabel('$x_1$', loc='right')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.plot(x, budsjett(2000, 25, 40, x), label = 'Konsumentens budsjett')
```
